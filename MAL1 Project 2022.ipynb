__(b)__ Based on the answers from all 83 candidates for the Horsens city council, perform a Principal Component Analysis with 2 principal components. Plot the results in a figure using these 2 components as the axes. Label the points with the party letter and the appropriate color.

```python
df_horsens = pd.DataFrame(data= horsens)
horsens_pca18 = decomposition.PCA(n_components=18)
horsens_res18 = horsens_pca18.fit_transform(df_horsens)
ratio = horsens_pca18.explained_variance_ratio_


# visualizing
print(ratio)
df_horsens_pca18 = pd.DataFrame(data=ratio,
                                columns=["ratio"])
sns.barplot(x = df_horsens_pca18.index.values,y = "ratio",data = df_horsens_pca18)
plt.show()
```

```python
# Plotting with matplotlib
# PCA with 2 Dimensions
df_horsens_party = pd.DataFrame(data= horsens_party,
                                columns=["Party"],)

parties_label = ["A","B","C","D","F","I","O","V","Z"]
party_colors = ["Red", "Indigo", "Green", "Black", "Fuchsia", "Cyan","Yellow","Blue", "#8B0000"]

def get_color_list(dataframe):
    color_list = []
    for party in dataframe["Party"]:
        for label in parties_label:
            if party == label:
                color_list.append(party_colors[parties_label.index(label)])
    return color_list

def get_color_number(dataframe):
    color_number = []
    for party in dataframe["Party"]:
        for label in parties_label:
            if party == label:
                color_number.append(parties_label.index(label))
    return color_number


color_list = get_color_list(df_horsens_party)
color_number = get_color_number(df_horsens_party)
```

```python
# build a PCA model
pca = decomposition.PCA(n_components=2)

pca.fit(df_horsens)
# transform the digits data onto the first two principal components
df_horsens_pca = pca.transform(df_horsens)

fig = plt.figure(figsize=(10, 10))

plt.figure(figsize=(10, 10))
plt.xlim(df_horsens_pca[:, 0].min(), df_horsens_pca[:, 0].max())
plt.ylim(df_horsens_pca[:, 1].min(), df_horsens_pca[:, 1].max())
for i in range(len(df_horsens)):
    # actually plot the digits as text instead of using scatter
    plt.text(df_horsens_pca[i, 0], df_horsens_pca[i, 1], str(df_horsens_party["Party"][i]),
             color = party_colors[color_number[i]],
             fontdict={'weight': 'bold', 'size': 9})
plt.xlabel("First principal component")
plt.ylabel("Second principal component")

# reversing x-axis to have a better view of parties and their alignings
plt.gca().invert_xaxis()
```

Confident that we can predict the partisan affiliation of a politician reasonably well based on their answers to the test, we want to build a model that will allow us to distinguish between the 9 political parties. For this purpose, we split the data into a training and a validation set.

__(e)__ Split the data into a training and a validation set, with appropriate fractions.

```python
from sklearn.model_selection import train_test_split

X_trainval, X_test, y_trainval, y_test = train_test_split(df_databorg.values, df_databorg_party["Party"], random_state=69)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, random_state=69)
print("Size of training set:{}".format(X_train.shape[0]))
print("Size of test set:{}".format(X_test.shape[0]))
```

First, we assume that a Naive Bayes approach is sufficient for our purposes.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (confusion_matrix,precision_score,recall_score,f1_score,
                             roc_curve,roc_auc_score,precision_recall_curve,accuracy_score,classification_report)

model = GaussianNB()

model.fit(X_trainval, y_trainval)
print("Accuracy score on training/validation set: {}".format(model.score(X_trainval, y_trainval)))
print("Accuracy score on test set: {}".format(model.score(X_test, y_test)))


predictions = model.predict(X_test)
print("Confusion matrix:  \n{}\n".format(confusion_matrix(y_test,predictions)))
print("Accuracy: \n{}\n".format(model.score(X_test,y_test)))
print("Precision: \n{}\n".format(precision_score(y_test,predictions,pos_label=1, average=None)))
print("Recall: \n{}\n".format(recall_score(y_test,predictions,pos_label=1, average=None)))
print("F1: \n{}".format(f1_score(y_test,predictions,pos_label=1,average=None)))
```

__(i)__ Play around with different values of _k_. Decide on a "good" value of _k_. Comment on the results.

```python
from sklearn.model_selection import cross_val_score

best_neighbour_score = 0
for num_neighbors in range(1,20):
    # Set a certain number of neighbors
    knn = KNeighborsClassifier(n_neighbors=num_neighbors)

    # Perform cross validation
    scores = cross_val_score(knn, X_trainval, y_trainval, cv=5)

    # Compute the mean score
    score = scores.mean()

    # If improvement, store score and parameter
    if score>best_neighbour_score:
        best_neighbour_score = score
        best_num_neighbors = num_neighbors

# Build a model on the combine training and valiation data
knn = KNeighborsClassifier(n_neighbors=best_num_neighbors)
knn.fit(X_trainval, y_trainval)

print("Best number of neighbors found: {}".format(best_num_neighbors))
print("Best average score: {}".format(best_neighbour_score))
print("Score on training/validation set: {}".format(knn.score(X_trainval, y_trainval)))
print("Score on test set: {}".format(knn.score(X_test, y_test)))
print("Confusion matrix:  \n{}\n".format(confusion_matrix(y_test,knn.predict(X_test))))
```

We found that the best number of neighbours is 12. Score on training is slightly lower than when Hyperparameter is set to 5, but  test result has increased with 1%.

Precision shows the ratio of relevant instances among retrieved ones. The best is of KNN with forest not so behind.

```python
print("Naive Bayes:")
print("Recall: {}\n".format(recall_score(y_test,predictions,pos_label=1, average="macro")))

print("KNN:")
print("Recall: {}\n".format(recall_score(y_test,knn.predict(X_test),pos_label=1, average="macro")))

print("Forest:")
print("Recall: {}\n".format(recall_score(y_test,forest.predict(X_test),pos_label=1, average="macro")))
```

Recall shows us the TP/(TP+FN). The best score still has KNN.

```python
print("Naive Bayes:")
print("F1: {}\n".format(f1_score(y_test,predictions,pos_label=1,average="macro")))

print("KNN:")
print("F1: {}\n".format(f1_score(y_test,knn.predict(X_test),pos_label=1,average="macro")))

print("Forest:")
print("F1: {}\n".format(f1_score(y_test,forest.predict(X_test),pos_label=1,average="macro")))
```

F1 is the function of Precision and Recall. F1 Score allows us to better measure the balance between Precision and Recall when there is uneven class distribution. The best F1 score still has KNN.
